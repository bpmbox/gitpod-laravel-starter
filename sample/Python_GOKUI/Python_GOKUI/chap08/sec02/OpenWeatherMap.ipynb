```python
# 現在の東京都の天気を取得する
import requests
import pprint

# APIキー
API_KEY = 'APIキー'
# OpenWeatherMapのAPIのURL
URL = 'http://api.openweathermap.org/data/2.5/'
# 地域を指定
place = 'Tokyo'
# 現在の天気を取得
data = requests.get(URL + 'weather?lang=ja&q=' + place + '&appid=' + API_KEY).json()
pprint.pprint(data)
```

```python
# JSONデータのキーを取得
data.keys()
```

```python
# 現在の天候を出力
data['weather'][0]['description']
```

```python
# 東京都の5日/ 3時間ごとの気象データを取得

# APIキー
API_KEY = "APIキー"
# OpenWeatherMapのURL
URL = "http://api.openweathermap.org/data/2.5/"
# 気象データを取得
data = requests.get(URL + 'forecast?q=Tokyo&lang=ja&appid=' + API_KEY).json()
pprint.pprint(data)
```

```python
# JSONデータの辞書のキーを取得
data.keys()
```

```python
# JSONデータの'list'キーの値(リスト)に
# インデックスの0から3時間毎の予報データが順番に格納されている
for i in data['list']:
    # 'dt_txt'キーで予報対象の日時を取得
    print(i['dt_txt'])
    # 'weather'キーの値はリストで、インデックス0の
    # 'description'キーに天候を示すデータが格納されている
    print(i['weather'][0]['description'])
```
